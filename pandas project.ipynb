```python
import pandas as pd
import seaborn as sb
import pandasql as pds
import matplotlib.pyplot as plt
```

```python
#I found  a dataset named " Global data on sustinable energy " that  shows data  about different types of energy from 2000 until 2020 for all the countries 
df=pd.read_csv('C:\\Users\\Home\\Phyton\\Datasets\\global-data-on-sustainable-energy (1).csv')
```

```python
#Entity: The name of the country or region .
#Year: The year for which the data is reported, from 2000 to 2020.
#Access to electricity (% of population): The percentage of population with access to electricity.
#Access to clean fuels for cooking (% of population): The percentage of the population with primary reliance on clean fuels.
#Renewable-electricity-generating-capacity-per-capita: Installed Renewable energy capacity per person
#Financial flows to developing countries (US $): Aid and assistance from developed countries for clean energy projects.
#Renewable energy share in total final energy consumption (%): Percentage of renewable energy in final energy consumption.
#Electricity from fossil fuels (TWh): Electricity generated from fossil fuels (coal, oil, gas) in terawatt-hours.
#Electricity from nuclear (TWh): Electricity generated from nuclear power in terawatt-hours.
#Electricity from renewables (TWh): Electricity generated from renewable sources (hydro, solar, wind, etc.) in terawatt-hours.
#Low-carbon electricity (% electricity): Percentage of electricity from low-carbon sources (nuclear and renewables).
#Primary energy consumption per capita (kWh/person): Energy consumption per person in kilowatt-hours.
#Energy intensity level of primary energy (MJ/$2011 PPP GDP): Energy use per unit of GDP at purchasing power parity.
#Value_co2_emissions (metric tons per capita): Carbon dioxide emissions per person in metric tons.
#Renewables (% equivalent primary energy): Equivalent primary energy that is derived from renewable sources.
#GDP growth (annual %): Annual GDP growth rate based on constant local currency.
#GDP per capita: Gross domestic product per person.
#Density (P/Km2): Population density in persons per square kilometer.
#Land Area (Km2): Total land area in square kilometers.
#Latitude: Latitude of the country.
#Longitude: Longitude of the country.

df
```

```python
#I cleaned the data by droping the columns that have more then 10% nulls 
# I also removed  the columns that refer to latitude and longitude( are not relevant in my analysis)
clean_df=df.drop(columns = ['Renewable-electricity-generating-capacity-per-capita','Financial flows to developing countries (US $)','Value_co2_emissions_kt_by_country' ,'Renewables (% equivalent primary energy)','Latitude','Longitude'])
```

```python
#I searched for the mean value in the Gross domestic product per person  and filled the nulls with this value

mean=clean_df['gdp_per_capita'].mean()
clean_df['gdp_per_capita'].fillna(mean,inplace=True)
```

```python
clean_df.corr(numeric_only=True)
```

```python
#I built a heatmap to see if there is a correlation between a high Gdp per capita(Gross product per person) and renewable energy .
#I found that sadly the richest countries do not use renewable energy.
#The only correlations that we can find is that in countries with a high Gdp per capita  we have a high Primary energy consumption and a relatively high acces to electricity and to lean fuels for cooking 

plt.figure(figsize=(12, 10))
sb.heatmap(clean_df.corr(numeric_only=True),cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# I found the median of Land area and replaced the missing value with the median value
medi=clean_df['Land Area(Km2)'].median()
clean_df['Land Area(Km2)'].fillna(medi,inplace=True)
```

```python
can=pds.sqldf("""select * from clean_df where Entity="Canada" ;""",globals())
usa=clean_df[clean_df['Entity']=='United States']
chn=pds.sqldf("""select * from clean_df where Entity="China" ;""",globals())
bra=pds.sqldf("""select * from clean_df where Entity="Brazil" ;""",globals())
aus=clean_df[clean_df['Entity']=='Australia']
```

```python
#After finding tre countries with the biggest land area I wanted to see over the years the progress in using renewable energy 
#The graph shows the evolution over the years and we can observe that the biggest growth is in China and in the USA 
#The other 3 big countries despite what we expect have no visible grown over the last 20 years in using less harming sources of energy

plt.figure(figsize=(12,10))
plt.title('Top 5 Countries by Land - Electricity from Renewables', fontsize=20, color='purple')
plt.xlabel('Year', color='green', fontsize=18)
plt.ylabel('Electricity from renewables (TWh)', color='green', fontsize=18)
sb.pointplot(x='Year',y='Electricity from renewables (TWh)', data=can,color='#E59214',label='Canada') 
sb.pointplot(x='Year',y='Electricity from renewables (TWh)', data=usa,color='blue',label='USA')
sb.pointplot(x='Year',y='Electricity from renewables (TWh)', data=chn,color='red',label='China')
sb.pointplot(x='Year',y='Electricity from renewables (TWh)', data=bra,color='#5BDA5F',label='Brazil')
sb.pointplot(x='Year',y='Electricity from renewables (TWh)', data=aus,color='#C677EE',label='Australia')
plt.legend(title="Entity")
plt.show()
```

```python
# I wanted to find the top 20 countries with the highest gross domestic product so first Iremoved the empty values
#Then I calculate the mean gdp and finnaly I sorted the values in descending order and select the top 20
df_filtered1 = clean_df[(clean_df['Year'] >= 2000) & (clean_df['Year'] <= 2020)].dropna(subset=['gdp_per_capita'])
gdp_by_country = df_filtered1.groupby('Entity')['gdp_per_capita'].mean()
```

```python
top_gdp_countries = gdp_by_country.sort_values(ascending=False).head(20)
top_gdp_countries
```

```python
# I wanted to find the top 20 countries with the highest Electricity from nuclear (TWh so first Iremoved the empty values
#Then I calculate the mean Electricity from nuclear (TWh) and finnaly I sorted the values in descending order and select the top 20
df_filtered2 = clean_df.dropna(subset=['Electricity from nuclear (TWh)'])
df_max_nuclear = df_filtered2.groupby('Entity')['Electricity from nuclear (TWh)'].mean()
top_nuclear_countries = df_max_nuclear.sort_values(ascending=False).head(20)
```

```python
top_nuclear_countries
```

```python
# I wanted to show how many of the top 20 gdp countries are in the top 20 nuclear energy countries and I found that there are 8 common countries
common_countries = set(top_gdp_countries.index).intersection(set(top_nuclear_countries.index))
common_countries_list = list(common_countries)
common_countries_series = pd.Series(common_countries_list)
```
